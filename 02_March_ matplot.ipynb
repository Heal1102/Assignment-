# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

```python
###Matplotlib is a popular Python library used for creating static, animated, and interactive visualizations 
#in Python. It provides a wide range of plotting functionalities for creating high-quality charts,
#graphs, and other visual representations of data. Matplotlib is extensively used in fields such 
#as data science, machine learning, finance, biology, and many others.

####Some common use cases of Matplotlib include:

# 1.Data exploration and analysis: Matplotlib provides a convenient way to visualize data to gain insights and understand patterns.

# 2.Presentation and reporting: Matplotlib allows users to create visually appealing plots and charts to communicate findings effectively.

# 3.Model evaluation: Visualizing model performance metrics such as accuracy, precision, recall, and ROC curves can help in evaluating and comparing different machine learning models.

# 4.Publication-quality figures: Matplotlib produces high-quality figures suitable for publication in academic papers, reports, and presentations.

# 5.Custom visualization: Matplotlib provides flexibility for creating custom visualizations tailored to specific requirements and preferences.

###Pyplot is a submodule of Matplotlib that provides a MATLAB-like interface for creating plots. Some common plots that can be created using the Pyplot module include:

# 1.Line plot: A basic plot used to display data points connected by straight lines.

# 2.Scatter plot: A plot used to visualize the relationship between two variables by displaying individual data points.

# 3.Histogram: A graphical representation of the distribution of numerical data, where data is divided into bins and the frequency of each bin is plotted.

# 4.Bar plot: A plot used to represent categorical data with rectangular bars, where the length of each bar corresponds to the value of the data.

# 5.Pie chart: A circular statistical graphic divided into slices to illustrate numerical proportions, where each slice represents a proportion of the whole.




```

# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
 plot a scatter plot.
 import numpy as np
 np.random.seed(3)
 x = 3 + np.random.normal(0, 2, 50)
 y = 3 + np.random.normal(0, 2, len(x))
 Note: Also add title, xlabel, and ylabel to the plot.

```python
## A scatter plot is a type of data visualization that represents individual data points
#on a two-dimensional plane.
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.figure(figsize=(10,6))
plt.title("Random data")
plt.xlabel("X label data")
plt.ylabel("Y label data")
plt.scatter(x,y)
plt.show
```
